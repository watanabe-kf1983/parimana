```python
import pandas as pd
def pivot_ab(sr):
    return pd.pivot_table(sr.to_frame(), index="a", columns="b")
```

```python
from parimana.settings import Settings
```

```python
settings=Settings(race_type="KEIBA")
settings
```

```python
race=settings.race
race
```

```python
odds = race.get_odds()
odds
```

```python
dist = race.destribution()
```

```python
import parimana.analyse.win_rate as wr

rel = dist.relations
mems = dist.members
win_rates = wr.extract_win_rate(rel, mems)
wr_df = wr.df_from_win_rate(win_rates)

pivot_ab(wr_df["win_rate"])
```

```python
pivot_ab(wr_df["win_rate_acc"])
```

```python
acc = wr_df["win_rate_acc"].groupby("a").sum()
acc.to_frame()
```

```python
acc_sum = acc.sum()
acc_sum
```

```python
from parimana.analyse.correlation import cor_none, cor_mapping_to_sr 

cor = cor_none(mems)
cor_sr = cor_mapping_to_sr(cor)
pivot_ab(cor_sr)
```

```python
from parimana.analyse.ability import (
    estimate_ability_map,
    find_uncertainty_map,
)
corwr_df = cor_sr.to_frame().join(wr_df)
u_map = find_uncertainty_map(corwr_df)
```
